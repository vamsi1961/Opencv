```
import cv2
from google.colab.patches import cv2_imshow
```

```
image_back=cv2.imread("/content/image_back.png")
image_circle=cv2.imread("/content/image_circle.png")
cv2_imshow(image_back)
```

```
cv2_imshow(image_circle)
```

```
image_and=cv2.bitwise_and(image_back,image_circle,mask=None)
cv2_imshow(image_and)
```

```
image_or=cv2.bitwise_or(image_back,image_circle,mask=None)
cv2_imshow(image_or)
```

```
image_not_circle=cv2.bitwise_not(image_circle,mask=None)
cv2_imshow(image_not_circle)
```

```
bottle=cv2.imread("/content/bottle.jpg")
bottle=cv2.resize(bottle,(500,500))
cv2_imshow(bottle)
```

```
average_blur=cv2.blur(bottle,(9,9))
cv2_imshow(average_blur)
```

```
gauss_noise=cv2.imread("/content/bottle_gaussian.jpg")
gauss_noise=cv2.resize(gauss_noise,(500,500))
cv2_imshow(gauss_noise)
```

```
bottle_gauss=cv2.GaussianBlur(gauss_noise,(7,7),0)
cv2_imshow(bottle_gauss)
```

```
bottle_median=cv2.imread("/content/bottle_median.jpg")
bottle_median=cv2.resize(bottle_median,(500,500))
cv2_imshow(bottle_median)
```

```
bottle_median=cv2.medianBlur(bottle_median,7)
cv2_imshow(bottle_median)
```

```
import numpy as np
kernel=(1/9)*np.array([[1,1,1],[1,1,1],[1,1,1]])
average=cv2.filter2D(bottle,-1,kernel)
cv2_imshow(average)
```
