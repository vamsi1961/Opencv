```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
bottle=cv2.imread("/content/bottle.jpg")
bottle=cv2.resize(bottle,(500,500))
cv2_imshow(bottle)
```

```python
import numpy as np
kernel=(1/9)*np.array([[1,1,1],[1,1,1],[1,1,1]])
average=cv2.filter2D(bottle,-1,kernel)
cv2_imshow(average)
```
